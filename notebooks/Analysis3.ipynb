```python
import pandas as pd
import numpy as mp
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

def eda(dframe):
    #check that input is pandas dataframe
    if type(dframe) != pd.core.frame.DataFrame:
        raise TypeError("Only pandas dataframe is allowed as input")
        
    #replace field thats entirely space(or empty) with NaN
    data = dframe.replace(r'^\s*$', np.nan, regex=True)
    
    print("Preview of data: ")
    display(dframe.head(3))
    
    print("\nTo check: \n (1) Total number of entries \n (2) Column types \n (3) Any null values\n")
    print(dframe.info())
    
    
    # generate preview of entries with null values
    if data.isnull().any(axis=None):
        print("\nPreview of data with null values:")
        display(data[data.isnull().any(axis=1)].head(3))
        
        plt.show()

    # generate count statistics of duplicate entries
    if len(data[data.duplicated()]) > 0:
        print("\nNumber of duplicated entries: ", len(data[data.duplicated()]))
        display(data[data.duplicated(keep=False)].sort_values(by=list(data.columns)).head())
    else:
        print("\nNo duplicated entries found")
        
```
