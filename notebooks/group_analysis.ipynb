### To find out which airline had the maximum fatilites I took the top 10 most crashed airline operators and then plotted a bar chart that gave the following conclusion 

```python
total_death_per_airline = df.groupby('Operator')[['Fatilities']].sum() 
total_death_per_airline = df_death_airline.rename(columns={"Operator": "Total Death"})
total_death_per_airline = df_death_airline.sort_values(by='Fatilities', ascending=False)
total_death_per_airline_top10 = df_death_airline.head(10)

total_death_per_airline_top10
```

```python

plt.figure(figsize=(15, 5))
sns.barplot(y=total_death_per_airline_top10.index, x="Fatilities", data=total_death_per_airline_top10, palette="pastel")
plt.title('Total deaths vs operator')
plt.xlabel('Fatilities')
plt.ylabel('Operator')
plt.show()
```

The maximum number of fatilities were incurred to airfrance with fatilites being 200.
